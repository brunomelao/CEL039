```python
#Programe aqui sua resposta da 1

import numpy as np
from control.matlab import *
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly
num = [1]
den = [1, 0.2, 1, 1]
ft= tf(num,den)
print(ft)
w=np.logspace(-2,2,500)
fig=make_subplots(rows=1,cols=1)

real, imag, freq = nyquist(ft,w, plot=False)

fig.add_trace(go.Scatter(
            x=real,
            y=imag,
            mode = 'lines',
            hoverinfo='all',
            legendgroup= 'Nyquist',
            line_color= 'black',
            name= 'Nyquist',
            showlegend= True), row=1, col=1)
fig.update_yaxes(title_text="$Im\{G(j\omega)\}$", range=[-2, 2], autorange=False,
       zeroline=True,
       showline=True,
       gridwidth=2,
       zerolinecolor='#969696',
       zerolinewidth=2,
       linewidth=6, row=1, col=1)
  
fig.update_xaxes(title_text="$Re\{G(j\omega)\}$", range=[-2,2 ], autorange=False,        
       zeroline=True,
       showline=True,
       gridwidth=2,
       zerolinecolor='#969696',
       zerolinewidth=2,
       linewidth=6, row=1, col=1)

fig.show()

roots = np.roots(den)

# Imprimir as raízes
print("Raízes:", roots)
```
